# Overdoses in Virginia

```python
import matplotlib.pyplot as plt
import pandas as pd

""" Input data """
df_op = pd.read_csv('C:/Users/Michael C. Springer/Desktop/William & Mary/Competing/M2/OpiodsVA.csv')
df_pres = pd.read_csv('C:/Users/Michael C. Springer/Desktop/William & Mary/Competing/M2/presidents.csv')
df_nsa = pd.read_csv('C:/Users/Michael C. Springer/Desktop/William & Mary/Competing/M2/TOTALNSA.csv')
```

### Opiod Overdoses in Virgina (per 100,000 residents) as compared to Median Household Income

```python
#Opiod versus Median Income
""" create variables for accessing the Figure adn Axes sub-objects """
fig,ax = plt.subplots()

""" Axes sub-object properties and methods """

ax.scatter(df_op['FPOO-Rate'],df_op['MedianHouseholdIncome'], alpha=0.5,s=75)
ax.xaxis.set_label_text('Opiod Overdoses (per 100,000 residents)', fontsize = 16)
ax.yaxis.set_label_text('Median Household Income', fontsize = 16)
ax.xaxis.set_tick_params(labelsize=14)
ax.yaxis.set_tick_params(labelsize=14)

""" Figure sub-object properties and methods """
fig.suptitle('Opiod Overdoses Correlated to Income', fontsize = 20)
fig.set_size_inches(10,7)
fig.savefig('opiod.jpg')

""" Display first graph """
plt.show()
```

#### Do opioid overdoes tend to be associated with less affluent areas—that is, areas where families have lower incomes?

The answer to this is complicated but, according to the data, there seems to be some correlation of median household income to opiod overdoses in Virginia where a the most overdoses did occur in the lower median income bands but this doesn't tell a complete story. Many counties in that same income bracket also reported zero overdoses.
